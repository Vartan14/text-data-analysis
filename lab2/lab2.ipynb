**1. Робота з text1**

```python
# Зчитування файлу
with open('text1.txt', 'r') as file:
    text = file.read()
```

```python
text
```

```python
# Імпорт бібліотеки та завантаження даних
import nltk
nltk.download('all')
```

```python
# a) Порахувати кількість речень в тексті і вивести останнє

# Токенізація речень
sentences = nltk.sent_tokenize(text)

num_sentences = len(sentences)
last_sentence = sentences[-1]

print(f"Кількість речень в тексті: {num_sentences}")
print(f"Останнє речення в тексті: {last_sentence}")
```

```python
# б) Позначити частини мови

# Токенізація слів
words = nltk.word_tokenize(text)

# Отримання частин мови
pos_tags = nltk.pos_tag(words)

pos_tags
```

```python
# в) Знайти 10 слів, які зустрічаються найчастіше
from collections import Counter
import re

# Токенізація слів та приведення їх до нижнього регістру
words = nltk.word_tokenize(text.lower())

# Видалення токенів, що не є словами
words = [word for word in words if re.match('^[a-zA-Z]+$', word)]

# Підрахунок слів
word_freq = Counter(words)

# Виділення 10-ти найпопулярніших слів
top_10_words = word_freq.most_common(10)

print(f'10 слів, які зустрічаються найчастіше:\n{top_10_words}')
```

**2. Робота з корпусом Brown**

```python
# Завантаження корпусу Brown та вибір категорії science_fiction
from nltk.corpus import brown

nltk.download('brown')
science_fiction_words = brown.words(categories='science_fiction')
```

```python
# a) Порахувати загальну кількість слів в категорії

print(f"Загальна кількість слів в категорії 'science_fiction': {len(science_fiction_words)}")
```

```python
# б) Видалити всі дієслова в другому тексті

# Об'єднання слів у текст
science_fiction_text = " ".join(science_fiction_words)

science_fiction_text
```

```python
# Розбиваємо текст на слова та визначаємо частини мови
tokens = nltk.word_tokenize(science_fiction_text)
pos_tags = nltk.pos_tag(tokens)

pos_tags[:100]
```

```python
# Теги, що відповідають дієсловам
verbs = ['VBP', 'VBN', 'VBG', 'VBD', 'VB', 'M']

# Вибираємо лише слова, які не є дієсловами
filtered_words = [word for word, pos in pos_tags if pos not in verbs]

print(filtered_words)
```

```python
# Об'єднаємо список у текст
text_without_verbs =  ' '.join(filtered_words)
text_without_verbs
```
